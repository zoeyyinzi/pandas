### Question 1. List all International Cities July Fares in USD (Use "Fares Data" and "FX Rates")

#### Read the tabs "Fares Data" and "FX Rates" from the excel file

```python
fares = pd.read_excel("Revenue_Accountant_Exercise_Staff_Senior_ZoeyZhou.xlsx","Fares Data")
fares.head()
```

```python
fx = pd.read_excel("Revenue_Accountant_Exercise_Staff_Senior_ZoeyZhou.xlsx","FX Rates")
fx.head()
```

#### Filter tab "Fares Data" where Dates in July and Country is International

```python
fares['Date'] = pd.to_datetime(fares['Date'])
fares['Date']

filtered_fares = fares.loc[(fares['Date'] >= '2019-07-01') & (fares['Country'] == 'International')]
filtered_fares
```

#### Join the two tables "Fares Data" and "FX Rates" to find the associated July 2019 Rate

```python
join =  pd.merge(filtered_fares,
                    fx,
                    left_on='Currency Code', right_on='Currency')
join
```

#### Add the Column "July Fares in USD" 

```python
join['July Fares in USD'] = join['Fares Amount'] / join['July 2019 Rate']
join = join[["City", "July Fares in USD"]]
join
```

#### Answer to Question 1 - See below

```python
question1 = join.groupby('City').sum()
question1['July Fares in USD'] = question1['July Fares in USD'].astype('int32')
question1
```

### Question 2. Show Month over Month (June vs July) $ and % Change in Driver Incentives for all cities. (Use "Driver Incentives Data")

```python
incentives = pd.read_excel("Revenue_Accountant_Exercise_Staff_Senior_ZoeyZhou.xlsx","Driver Incentives Data")
incentives.head()
```

```python
incentives['Month'] = pd.DatetimeIndex(incentives['Date']).month
incentives.head()
```

```python
table = pd.pivot_table(incentives, values='Amount USD', index=['City'],
                    columns=['Month'], aggfunc=sum)
table
```

#### Answer to Question 2 - See below

```python
table['$ Change'] = table[7] - table[6]
table['% Change'] = table['$ Change'] / table[6]
table
```
